## Selecting Elements Using the xs() Function in Multi-Indexed DataFrames

```python
import numpy as np
import pandas as pd
```

```python
inside = ["class A", "class B", "class C", "class A", "class B", "class C" ]
outside = ['school1', 'school1', 'school1', 'school2', 'school2', 'school2']

multi_index = list(zip(outside, inside))

hier_index = pd.MultiIndex.from_tuples(multi_index)

np.random.seed(101)
df = pd.DataFrame(np.random.randint(70, 100, size = (6, 2)), index = hier_index, columns = ["1st_semester", "2st_semester"])

df.index.names = ["schools", "classes"]
```

```python
df
```

```python
df.xs("school2")
```

```python
df.xs(("school2", "class A"))
```

```python
df
```

```python
df.xs(("school2", "class A"), level = [0, 1])
```

```python
df.xs("class A", level = "classes")
```

```python
df.xs("class A", level = 1)
```

## Joining Pandas Dataframes: Join() Function

```python
import pandas as pd
import numpy as np
```

```python
df1 = pd.DataFrame({'X': ['X0', 'X1', 'X2'],
                     'Y': ['Y0', 'Y1', 'Y2']},
                      index = ['A0', 'A1', 'A2']) 

df2 = pd.DataFrame({'Z': ['Z0', 'Z2', 'Z3'],
                    'T': ['T0', 'T2', 'T3']},
                      index = ['A0', 'A2', 'A3'])
```

```python
df1
```

```python
df2
```

```python
df1.join(df2)
```

```python
df1.join(df2, how = "inner")
```

```python
df1.join(df2, how = "outer")
```

```python
df1.join(df2, how = "right")
```

```python
df3 = pd.DataFrame({'key': ['X0', 'X2', 'X3', 'X4', 'X5', 'X6'],
                   'Y': ['Y0', 'Y2', 'Y3', 'Y4', 'Y5', 'Y6']})

df4 = pd.DataFrame({'key': ['X0', 'X2', 'X3'],
                      'Z': ['Z0', 'Z2', 'Z3']})
```

```python
df3
```

```python
df4
```

```python
df3.join(df4)
```

```python
df3.join(df4, lsuffix = "_left", rsuffix = "_right")
```

```python
df3.join(df4, lsuffix = "_left", rsuffix = "_right", how = "inner")
```

```python
df3
```

```python
df3.set_index("key")
```

```python
df3.set_index("key").join(df4.set_index("key"))
```
